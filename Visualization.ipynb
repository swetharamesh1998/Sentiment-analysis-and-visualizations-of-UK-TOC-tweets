```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('sentiment_data_with_predicted_sentiment.xlsx')


df_regions = df['Region'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Region')


region_tweet_count = df_regions.value_counts()


region_tweet_count.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Tweets by Region')
plt.xlabel('Region')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('sentiment_data_with_regions.xlsx')


df_regions = df['Region'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Region')

region_tweet_count = df_regions.value_counts()


plt.figure(figsize=(10, 6))
plt.pie(region_tweet_count, labels=region_tweet_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Tweets by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('sentiment_data_with_predicted_sentiment.xlsx')


sentiment_counts = df['predicted_sentiment'].value_counts()
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Distribution of Sentiments')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('sentiment_data_with_predicted_sentiment.xlsx')

plt.figure(figsize=(10, 6))
sns.countplot(x='TOC', hue='predicted_sentiment', data=df, palette='Set2')
plt.xlabel('TOC (Train Operating Company)')
plt.ylabel('Number of Tweets')
plt.title('Sentiment Distribution by TOC')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('sentiment_data_with_regions.xlsx')


toc_counts = df['TOC'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=toc_counts.index, y=toc_counts.values, palette='viridis')
plt.title('Distribution of Tweets Across TOCs')
plt.xlabel('TOC')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
